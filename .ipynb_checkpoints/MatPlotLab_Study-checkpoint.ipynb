```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x1, y1, x2, y2)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.title("Plot testing", fontdict = font1, loc = 'left')

#plt.grid()
plt.grid(axis = 'y', color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()

#plt.savefig(sys.stdout.buffer)
#plt.savefig("mpltlabtest.png")
#sys.stdout.flush()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode, shadow = True, colors = mycolors)
plt.legend(title = "Four Fruits:")
plt.show()
```

```python
# Instructions

# Write a Python program that generates a line plot using Matplotlib. The plot should display data points (1, 1), (2, 4), (3, 9), and (4, 16). 
# The title should be 'Simple Line Plot', the x-label should be 'X Axis', and the y-label should be 'Y Axis'. Ensure the plot is displayed when the program is run.
# Requirements

# The plot should have a title 'Simple Line Plot'.

# The x-axis label should be 'X Axis'.

# The y-axis label should be 'Y Axis'.
# The line plot should have the data points (1, 1), (2, 4), (3, 9), and (4, 16).

import matplotlib.pyplot as plt

def main():
    # Create data
    x = [1, 2, 3, 4]
    y = [1, 4, 9, 16]
    
    # Write your code here to create the plot
    plt.plot(x, y)
    plt.xlabel("X Axis")
    plt.ylabel("Y Axis")
    plt.title("Simple Line Plot")

    # Show the plot
    plt.show()

if __name__ == '__main__':
    main()
```
