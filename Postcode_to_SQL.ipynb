```python
import os
import pandas as pd
import zipfile
import pyarrow.parquet as pq
import pyarrow as pa

# Data for Phamacy registration
# https://www.pharmacyregulation.org/registers

# Data source from https://geoportal.statistics.gov.uk/datasets/265778cd85754b7e97f404a1c63aea04/about

curDir = os.getcwd()
#zf = zipfile.ZipFile(curDir + '/ONSPD_AUG_2024.zip')
zf = zipfile.ZipFile(curDir + '/ONSPD_FEB_2025.zip')
text_files = zf.infolist()

print ("Uncompressing and reading data... ")
#print(text_files)
```

```python
#import polars as pl
list_ = []
alist_ = []
clist_ = []
nlist_ = []
out = "combined.parquet"
writer = None
for text_file in text_files:
    #Define columns of the csv reading
    colhd = ['pcd','pcd2','pcds','dointr','doterm','oscty','ced','oslaua','osward','parish','usertype','oseast1m','osnrth1m','osgrdind','oshlthau','nhser','ctry',\
             'rgn','streg','pcon','eer','teclec','ttwa','pct','itl','statsward','oa01','casward','npark','lsoa01','msoa01','ur01ind','oac01','oa11','lsoa11','msoa11',\
             'wz11','sicbl','bua22','ru11ind','oac11','lat','long','lep1','lep2','pfa','imd','calncv','icb','oa21','lsoa21','msoa21']

    # Session of main csv files
    # print(text_file.filename)
    if text_file.filename.endswith("csv"):
        if text_file.filename.startswith("Data/multi_csv/"):
            # Start row in 2nd row of each csv as it is huge size. It could saving more memory in process rather than handle 1st row as header.
            df = pd.read_csv(zf.open(text_file.filename), skiprows=[0], names=colhd, dtype=str)
            list_.append(df)
            
df = pd.concat(list_)
```
