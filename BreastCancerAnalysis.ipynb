```python
#Breast Cancer Analysis
import os
os.environ["PYARROW_IGNORE_TIMEZONE"] = "1"
import numpy as np
import pandas as pd
```

```python
bcancer= pd.read_csv('breast-cancer-wisconsin_wdbc.csv')
```

```python
bcancer.info()
```

```python
#bcancer.describe()
#bcancer.shape
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

x = np.array(bcancer[['radius (nucC)']][bcancer['diagnosis (M=malignant; B=benign)']=='B'])
y = np.array(bcancer[['fractal dimension (nucC)']][bcancer['diagnosis (M=malignant; B=benign)']=='B'])
#sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("radius (nucC)", fontdict = font2)
plt.ylabel("fractal dimension (nucC)", fontdict = font2)
plt.title("Radius vs fractal dimension (benign)", fontdict = font1)

plt.scatter(bcancer[['radius (nucC)']][bcancer['diagnosis (M=malignant; B=benign)']=='B'], bcancer[['fractal dimension (nucC)']][bcancer['diagnosis (M=malignant; B=benign)']=='B'])
plt.show()
```

```python
x = np.array(bcancer[['radius (nucC)']][bcancer['diagnosis (M=malignant; B=benign)']!='B'])
y = np.array(bcancer[['fractal dimension (nucC)']][bcancer['diagnosis (M=malignant; B=benign)']!='B'])

plt.xlabel("radius (nucC)", fontdict = font2)
plt.ylabel("fractal dimension (nucC)", fontdict = font2)
plt.title("Radius vs fractal dimension (malignant)", fontdict = font1)

plt.scatter(bcancer[['radius (nucC)']][bcancer['diagnosis (M=malignant; B=benign)']!='B'], bcancer[['fractal dimension (nucC)']][bcancer['diagnosis (M=malignant; B=benign)']!='B'])
plt.show() 
```

```python
## R-Squared : Checking any relationship between x and y
# if the value become 1 means having relationship
import numpy
from sklearn.metrics import r2_score

x1 = [int(i*100) for i in x]
y1 = [int(i*100) for i in y]
#print(x1)
#print(y1)
mymodel = numpy.poly1d(numpy.polyfit(x1, y1, 3))

print(r2_score(y1, mymodel(x1))) 
```
